```python
# 3. Exploratory Data Analysis
df["Date"] = pd.to_datetime(df["Date"])
df.dtypes
```

```python
# profit margine %
df["Profit Margine"] = df["Profit"]/df["Sale Price"]*100
#sales values
df["Sales Value"] = df["Quantity"]* df["Sale Price"]
# cost values
df["Cost Value"]  = df["Quantity"]*df["Cost"]
df.head()
```
